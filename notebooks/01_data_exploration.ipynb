```python
# Use only training matches to compute team strength

train_df = df_clean[df_clean["year"] <= 2020].copy()

wins = {}

for _, row in train_df.iterrows():

    home_team = row["home_team"]
    away_team = row["away_team"]

    wins.setdefault(home_team, {"wins": 0, "matches": 0})
    wins.setdefault(away_team, {"wins": 0, "matches": 0})

    wins[home_team]["matches"] += 1
    wins[away_team]["matches"] += 1

    if row["result"] == "Home Win":
        wins[home_team]["wins"] += 1

    elif row["result"] == "Away Win":
        wins[away_team]["wins"] += 1


team_win_rate = {
    team: stats["wins"] / stats["matches"]
    for team, stats in wins.items()
}
```

```python
team_win_rate["Brazil"]

```

```python
team_win_rate["Morocco"]
```

```python
df_features = df_clean.copy()

df_features["home_team_win_rate"] = (
    df_features["home_team"]
    .map(team_win_rate)
    .fillna(0.5)
)

df_features["away_team_win_rate"] = (
    df_features["away_team"]
    .map(team_win_rate)
    .fillna(0.5)
)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# =========================
# 1. Compute average goals scored using TRAINING DATA ONLY
# =========================

train_df = df_clean[df_clean["year"] <= 2020].copy()

team_goals = {}

for _, row in train_df.iterrows():
    home_team = row["home_team"]
    away_team = row["away_team"]

    team_goals.setdefault(home_team, {"goals_scored": 0, "matches": 0})
    team_goals.setdefault(away_team, {"goals_scored": 0, "matches": 0})

    team_goals[home_team]["goals_scored"] += row["home_score"]
    team_goals[home_team]["matches"] += 1

    team_goals[away_team]["goals_scored"] += row["away_score"]
    team_goals[away_team]["matches"] += 1


team_avg_goals_scored = {
    team: stats["goals_scored"] / stats["matches"]
    for team, stats in team_goals.items()
}

# =========================
# 2. Add features
# =========================

df_features_goals = df_features.copy()

df_features_goals["home_team_avg_goals_scored"] = (
    df_features_goals["home_team"]
    .map(team_avg_goals_scored)
    .fillna(1.3)
)

df_features_goals["away_team_avg_goals_scored"] = (
    df_features_goals["away_team"]
    .map(team_avg_goals_scored)
    .fillna(1.3)
)

# =========================
# 3. Build feature matrix
# =========================

X_goals = df_features_goals[
    [
        "home_team",
        "away_team",
        "tournament",
        "country",
        "neutral",
        "year",
        "home_team_win_rate",
        "away_team_win_rate",
        "home_team_avg_goals_scored",
        "away_team_avg_goals_scored",
    ]
]

y_goals = df_features_goals["result"]

X_goals_encoded = pd.get_dummies(
    X_goals,
    columns=[
        "home_team",
        "away_team",
        "tournament",
        "country"
    ]
)

# =========================
# 4. Chronological split
# =========================

train_mask = df_features_goals["year"] <= 2020
test_mask = df_features_goals["year"] > 2020

X_train_goals = X_goals_encoded[train_mask]
X_test_goals = X_goals_encoded[test_mask]

y_train_goals = y_goals[train_mask]
y_test_goals = y_goals[test_mask]

# =========================
# 5. Train model
# =========================

rf_goals = RandomForestClassifier(random_state=42)

rf_goals.fit(X_train_goals, y_train_goals)

y_pred_goals = rf_goals.predict(X_test_goals)

accuracy_goals = accuracy_score(y_test_goals, y_pred_goals)

print("Accuracy WITH win-rate + avg goals scored:", accuracy_goals)
print(classification_report(y_test_goals, y_pred_goals))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# =========================
# 1. Compute avg goals conceded using TRAINING DATA ONLY
# =========================

train_df = df_clean[df_clean["year"] <= 2020].copy()

team_conceded = {}

for _, row in train_df.iterrows():
    home_team = row["home_team"]
    away_team = row["away_team"]

    team_conceded.setdefault(home_team, {"goals_conceded": 0, "matches": 0})
    team_conceded.setdefault(away_team, {"goals_conceded": 0, "matches": 0})

    # Home team conceded away_score
    team_conceded[home_team]["goals_conceded"] += row["away_score"]
    team_conceded[home_team]["matches"] += 1

    # Away team conceded home_score
    team_conceded[away_team]["goals_conceded"] += row["home_score"]
    team_conceded[away_team]["matches"] += 1


team_avg_goals_conceded = {
    team: stats["goals_conceded"] / stats["matches"]
    for team, stats in team_conceded.items()
}

# =========================
# 2. Add conceded features
# =========================

df_features_full = df_features_goals.copy()

df_features_full["home_team_avg_goals_conceded"] = (
    df_features_full["home_team"]
    .map(team_avg_goals_conceded)
    .fillna(1.3)
)

df_features_full["away_team_avg_goals_conceded"] = (
    df_features_full["away_team"]
    .map(team_avg_goals_conceded)
    .fillna(1.3)
)

# =========================
# 3. Build feature matrix
# =========================

X_full = df_features_full[
    [
        "home_team",
        "away_team",
        "tournament",
        "country",
        "neutral",
        "year",
        "home_team_win_rate",
        "away_team_win_rate",
        "home_team_avg_goals_scored",
        "away_team_avg_goals_scored",
        "home_team_avg_goals_conceded",
        "away_team_avg_goals_conceded",
    ]
]

y_full = df_features_full["result"]

X_full_encoded = pd.get_dummies(
    X_full,
    columns=[
        "home_team",
        "away_team",
        "tournament",
        "country"
    ]
)

# =========================
# 4. Chronological split
# =========================

train_mask = df_features_full["year"] <= 2020
test_mask = df_features_full["year"] > 2020

X_train_full = X_full_encoded[train_mask]
X_test_full = X_full_encoded[test_mask]

y_train_full = y_full[train_mask]
y_test_full = y_full[test_mask]

# =========================
# 5. Train model
# =========================

rf_full = RandomForestClassifier(random_state=42)

rf_full.fit(X_train_full, y_train_full)

y_pred_full = rf_full.predict(X_test_full)

accuracy_full = accuracy_score(y_test_full, y_pred_full)

print("Accuracy WITH win-rate + goals scored + goals conceded:", accuracy_full)
print(classification_report(y_test_full, y_pred_full))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# =========================
# 1. Add goal difference features
# =========================

df_features_gd = df_features_full.copy()

df_features_gd["home_team_goal_difference"] = (
    df_features_gd["home_team_avg_goals_scored"]
    - df_features_gd["home_team_avg_goals_conceded"]
)

df_features_gd["away_team_goal_difference"] = (
    df_features_gd["away_team_avg_goals_scored"]
    - df_features_gd["away_team_avg_goals_conceded"]
)

# =========================
# 2. Build feature matrix
# =========================

X_gd = df_features_gd[
    [
        "home_team",
        "away_team",
        "tournament",
        "country",
        "neutral",
        "year",
        "home_team_win_rate",
        "away_team_win_rate",
        "home_team_avg_goals_scored",
        "away_team_avg_goals_scored",
        "home_team_avg_goals_conceded",
        "away_team_avg_goals_conceded",
        "home_team_goal_difference",
        "away_team_goal_difference",
    ]
]

y_gd = df_features_gd["result"]

X_gd_encoded = pd.get_dummies(
    X_gd,
    columns=[
        "home_team",
        "away_team",
        "tournament",
        "country"
    ]
)

# =========================
# 3. Chronological split
# =========================

train_mask = df_features_gd["year"] <= 2020
test_mask = df_features_gd["year"] > 2020

X_train_gd = X_gd_encoded[train_mask]
X_test_gd = X_gd_encoded[test_mask]

y_train_gd = y_gd[train_mask]
y_test_gd = y_gd[test_mask]

# =========================
# 4. Train model
# =========================

rf_gd = RandomForestClassifier(random_state=42)

rf_gd.fit(X_train_gd, y_train_gd)

y_pred_gd = rf_gd.predict(X_test_gd)

accuracy_gd = accuracy_score(y_test_gd, y_pred_gd)

print("Accuracy WITH goal difference:", accuracy_gd)
print(classification_report(y_test_gd, y_pred_gd))
```

```python
from collections import defaultdict
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# ==================================================
# RECENT FORM (last 10 matches approximation)
# ==================================================

train_df = df_clean[df_clean["year"] <= 2020].copy()

team_results = defaultdict(list)

for _, row in train_df.iterrows():

    home_team = row["home_team"]
    away_team = row["away_team"]

    # Home team result
    if row["result"] == "Home Win":
        team_results[home_team].append(1)
        team_results[away_team].append(0)

    elif row["result"] == "Away Win":
        team_results[home_team].append(0)
        team_results[away_team].append(1)

    else:
        team_results[home_team].append(0.5)
        team_results[away_team].append(0.5)

# Average of last 10 matches
recent_form = {}

for team, results in team_results.items():

    last_10 = results[-10:]

    recent_form[team] = sum(last_10) / len(last_10)

# ==================================================
# ADD FEATURES
# ==================================================

df_recent = df_features_full.copy()

df_recent["home_recent_form"] = (
    df_recent["home_team"]
    .map(recent_form)
    .fillna(0.5)
)

df_recent["away_recent_form"] = (
    df_recent["away_team"]
    .map(recent_form)
    .fillna(0.5)
)

# ==================================================
# FEATURES
# ==================================================

X_recent = df_recent[
    [
        "home_team",
        "away_team",
        "tournament",
        "country",
        "neutral",
        "year",
        "home_team_win_rate",
        "away_team_win_rate",
        "home_team_avg_goals_scored",
        "away_team_avg_goals_scored",
        "home_team_avg_goals_conceded",
        "away_team_avg_goals_conceded",
        "home_recent_form",
        "away_recent_form",
    ]
]

y_recent = df_recent["result"]

X_recent_encoded = pd.get_dummies(
    X_recent,
    columns=[
        "home_team",
        "away_team",
        "tournament",
        "country"
    ]
)

# ==================================================
# SPLIT
# ==================================================

train_mask = df_recent["year"] <= 2020
test_mask = df_recent["year"] > 2020

X_train_recent = X_recent_encoded[train_mask]
X_test_recent = X_recent_encoded[test_mask]

y_train_recent = y_recent[train_mask]
y_test_recent = y_recent[test_mask]

# ==================================================
# MODEL
# ==================================================

rf_recent = RandomForestClassifier(random_state=42)

rf_recent.fit(X_train_recent, y_train_recent)

y_pred_recent = rf_recent.predict(X_test_recent)

accuracy_recent = accuracy_score(
    y_test_recent,
    y_pred_recent
)

print("Accuracy WITH recent form:", accuracy_recent)

print(
    classification_report(
        y_test_recent,
        y_pred_recent
    )
)
```
